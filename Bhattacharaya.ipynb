```python
import numpy as np
import h5py
def read_train_hdf5():
    images, labels = [], []
    # Open the HDF5 file
    file = h5py.File("/content/drive/MyDrive/Necklace_Gallery.hdf5", "r+")
    images = np.array(file["/train_img"]).astype("uint8")
    labels = np.array(file["/train_labels"]).astype("uint8")
    return images, labels

def read_query_hdf5():
    images, labels = [], []
    # Open the HDF5 file
    file = h5py.File("/content/drive/MyDrive/Necklace_Query.hdf5", "r+")
    images = np.array(file["/train_img"]).astype("uint8")
    labels = np.array(file["/train_labels"]).astype("uint8")
    return images, labels
```

```python
import cv2
def np_hist_to_cv(np_histogram_output):
  counts, bin_edges = np_histogram_output
  return counts.ravel().astype('float32')
def findDifference(f1, f2):
  h1 = np.histogram(f1)
  h2 = np.histogram(f2)
  val=cv2.compareHist(np_hist_to_cv(h1),np_hist_to_cv(h2),cv2.HISTCMP_BHATTACHARYYA)
  return val
```

```python
def findDist(query_img,all_image):
  res=[]
  for i in range(len(all_image)):
    dist=findDifference(query_img,all_image[i])
    res.append(dist)
  return res
```
